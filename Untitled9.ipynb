```python
from astropy.io import fits
import numpy as np
```

```python

fits_file = "0112550201/pps/P0112550201PNS003PIEVLI0000.FTZ"
with fits.open(fits_file) as hdul:
    events = hdul[1].data

pi = events['PI']

energy_kev = pi / 1000.0

soft_band = (energy_kev >= 0.5) & (energy_kev <= 2.0)
hard_band = (energy_kev > 2.0) & (energy_kev <= 10.0)

soft_counts = np.sum(soft_band)
hard_counts = np.sum(hard_band)
```

```python
with fits.open('0112550201/pps/P0112550201PNS003SRSPEC0001.FTZ') as hdul:
    spec_data = hdul[1].data
    channels = spec_data['CHANNEL']
    counts = spec_data['COUNTS']
```

```python
if (hard_counts + soft_counts) > 0:
    hr = (hard_counts - soft_counts) / (hard_counts + soft_counts)
else:
    hr = np.nan
print(f"Hardness Ratio: {hr:.3f}")
```

```python
with fits.open(fits_file) as hdul:
    header = hdul[0].header
    exposure = header.get('EXPOSURE', 1.0)
soft_rate = soft_counts / exposure
hard_rate = hard_counts / exposure
if (hard_rate + soft_rate) > 0:
    hr_rate = (hard_rate - soft_rate) / (hard_rate + soft_rate)
    print(f"Hardness Ratio (rate-based): {hr_rate:.3f}")
```

```python

from astroquery.simbad import Simbad
from astropy.table import Table

# Get all the NGC/IC objects
def query_deep_sky_objects(xrism):
    adql_query = f"""
    SELECT TOP 10000 main_id, otype, ra, dec
    FROM basic
    WHERE main_id LIKE 'NGC%' OR main_id LIKE 'IC%'
    ORDER BY main_id ASC
    """
    result = Simbad.query_tap(adql_query)
    return result

objects = query_deep_sky_objects('NGC 5033')
objects.write('simbad_objects.csv', format='csv', overwrite=True)
```

```python
import matplotlib.pyplot as plt

def plot_lightcurve(obsid):
    events = fits.open(f"{obsid}/pps/P0871020101PNS003PIEVLI0000.FTZ")[1].data
    events = events[events['PATTERN'] <= 4]
    time_bins = np.arange(0, events['TIME'].max(), 500)  # 500s bins
    counts, _ = np.histogram(events['TIME'], bins=time_bins)

    plt.step(time_bins[:-1], counts, where='mid', label=f"ObsID {obsid}")

def plot_Lightcurve(obs_id):
    events = fits.open(f"{obs_id}/pps/P0112550201PNS003PIEVLI0000.FTZ")[1].data
    events = events[events['PATTERN'] <= 4]
    time_bins = np.arange(0, events['TIME'].max(), 500)  # 500s bins
    counts, _ = np.histogram(events['TIME'], bins=time_bins)

    plt.step(time_bins[:-1], counts, where='mid', label=f"ObsID {obs_id}")

plt.figure(figsize=(10,5))
plot_Lightcurve("0112550201")  # Old observation
plot_lightcurve("0871020101")  # Recent observation
plt.xlabel("Time (s)"); plt.ylabel("Counts/500s"); plt.legend()
plt.title("NGC 5033 Light Curve Comparison")
plt.show()
```

```python
def calculate_hr(obsid):
    events = fits.open(f"{obsid}/pps/P0871020101PNS003PIEVLI0000.FTZ")[1].data
    soft = (events['PI'] > 500) & (events['PI'] < 2000)
    hard = (events['PI'] > 2000) & (events['PI'] < 10000)
    hr = (hard.sum() - soft.sum()) / (hard.sum() + soft.sum())
    return hr

def Calculate_hr(obs_id):
    events = fits.open(f"{obs_id}/pps/P0112550201PNS003PIEVLI0000.FTZ")[1].data
    soft = (events['PI'] > 500) & (events['PI'] < 2000)
    hard = (events['PI'] > 2000) & (events['PI'] < 10000)
    hr = (hard.sum() - soft.sum()) / (hard.sum() + soft.sum())
    return hr

hr_old = Calculate_hr("0112550201")
hr_new = calculate_hr("0871020101")

print(f"Hardness Ratio (2-10 keV/0.5-2 keV):")
print(f"2002 Observation: {hr_old:.2f}")
print(f"2020 Observation: {hr_new:.2f}")
```
